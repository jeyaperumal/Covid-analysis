# 1. Import pandas and read in the india_statewise.json file into a dataframe called covid.


```python
covid=pd.read_json('/Users/sutharsha/Desktop/Courses_and_Internship/NIELIT/india_statewise.json')
covid
```

# 2. Normalize the dataframe with ‘data’ as column and ‘statewise’ as row and rename it as df_india

```python
df_india = pd.io.json.json_normalize(covid['data']['statewise'])
```

```python
df_india
```

# 3. Now make ‘state’ column as reset


```python
df_india = df_india.set_index('state')
df_india
```

# 4. Now Identify a row which is not a valid state and drop it

```python
df_india.drop(["Lakshadweep"], inplace = True) 
df_india.drop(["State Unassigned"], inplace = True) 

```

```python
df_india
```

# 5. Calculate the Mortality rate (deaths/confirmed) for each state and identify top 5 sates with highest mortality rate.

```python
df_india['mortality rate'] = df_india['deaths'] / df_india['confirmed']
```
